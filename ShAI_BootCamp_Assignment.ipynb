```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load your dataset
df = pd.read_csv('Salaries.csv')
df.head(50)
```

```python
df.drop(["Notes","Status"],axis=1,inplace=True)
```

```python
df["JobTitle"].unique()
```

```python
df.columns
```

```python
print("Number of rows and columns:", df.shape[1])
```

```python
print("Number of rows and columns:", df.shape[0])
```

```python
print("\nData types of each column:\n", df.dtypes)
```

```python
print("\nMissing values in each column:\n", df.isnull().sum())
```

```python
df.describe()
```

```python
df['TotalPay'].fillna(df['TotalPay'].mean(), inplace=True)
df['BasePay'].fillna(0, inplace=True)
df['OvertimePay'].fillna(0, inplace=True)
df['OtherPay'].fillna(0, inplace=True)
df['Benefits'].fillna(0, inplace=True)
```

```python
print("\nMissing values in each column:\n", df.isnull().sum())
```

```python
unique_departments_count = df['JobTitle'].nunique()
unique_departments_count
```

```python
filtered_df = df[df['JobTitle'].str.contains('DEPARTMENT', case=False, na=False)]
filtered_df
```

```python
# Create histogram for salary distribution
plt.hist(df['TotalPay'], bins=20, edgecolor='black')
plt.title('TotalPay Distribution')
plt.xlabel('TotalPay')
plt.ylabel('Frequency')
plt.show()

# Create pie chart for department distribution
department_counts = df['JobTitle'].value_counts()
plt.pie(department_counts, labels=department_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('JobTitle Distribution')
```

```python
grouped_data = df.groupby('JobTitle')['TotalPay'].mean()
print("\nAverage Salary by Department:\n", grouped_data)
```

```python
# Identify correlation between salary and another numerical column (e.g., Years of Experience)
correlation = df['TotalPay'].corr(df['Year'])

# Scatter plot to visualize the relationship
plt.scatter(df['Year'], df['TotalPay'])
plt.title('Scatter Plot: Salary vs. Years of Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()

print("\nCorrelation between Salary and Years of Experience:", correlation)
```
